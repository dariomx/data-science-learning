# 1 
<b>If $A_n \subset A_{n-1} \subset \dots \subset A_1$ show that $\cap_{i=1}^n A_i = A_n$, $\cup_{i=1}^n A_i = A_1$ </b>

We will prove it by induction over $n$ each one of the results.

#### 1.1 $\cup_{i=1}^n A_i = A_n$

###### 1.1.1 Base case $n=2$

For $n=2$ the hypothesis simplifies to:
\begin{equation*}
A_2 \subset A_1
\end{equation*}

And the equality to prove simplifies to 

\begin{equation*}
A_1 \cap A_2 = A_2
\end{equation*}

In strict theory, for any set equality to be true, we would need to prove two things:

\begin{equation*}
(A_1 \cap A_2) \subset A_2 \;\land\; A_2 \subset (A_1 \cap A_2)
\end{equation*}

In other words, we need to prove that any element of $(A_1 \cap A_2)$ is also in $A_2$ and viceversa. Let us begin by proving the first part.

Let $x \in (A_1 \cap A_2)$, by definition of set intersection we know that:

\begin{equation*}
x \in A_1 \;\land\; x \in A_2
\end{equation*}

By the laws of logic, if the AND is true as a whole, we know that each individual operand is also true. In particular, we can tell that

\begin{equation*}
x \in A_2
\end{equation*}

But if the $x$ is also in $A_2$, then by definition of $\subset$ operator we can tell that $(A_1 \cap A_2) \subset A_2$; which is want we wanted to prove. 

Let us go now in the opposite direction, let us pick a $y \in A_2$ and prove that $y$ also belongs to $(A_1 \cap A_2)$. By hypothesis we know that $A_2 \subset A_1$, hence $y \in A_1$. So we have a couple of conditions holding now:

<ol>
<li> $y \in A_2$ </li>
<li> $y \in A_1$ </li>
</ol>

But if both conditions above hold, is the same as having an AND over them:

\begin{equation*}
y \in A_1 \;\land\; y \in A_2
\end{equation*}

And by definition of the intersection of sets, we know that $y \in (A_1 \cap A_2)$; which is the second part we needed to prove.

Putting together:

\begin{equation*}
(A_1 \cap A_2) \subset A_2 \;\land\; A_2 \subset (A_1 \cap A_2) 
\;\;\implies\;\; (A_1 \cap A_2) = A_2
\end{equation*}

and this proves the base case.

###### 1.1.2 Inductive case $P(n) \implies P(n+1)$

For the inductive case, we will prove that assuming the property true for $n$, implies the property also holds for $n+1$ (in other words, we want to prove that the property is "inherited"). 

Let us look first at the induction hypothesis $P(n)$:

\begin{equation*}
\cap_{i=1}^n A_i = A_n
\end{equation*}

and let us remember the additional hyphothesis for the exercise:

\begin{equation*}
A_{n+1} \subset A_n \subset A_{n-1} \subset \dots \subset A_1
\end{equation*}

using the two hypothesis above we need to prove the property $P(n+1)$, which is:

\begin{equation*}
\cap_{i=1}^{n+1} A_i = A_{n+1}
\end{equation*}

Let us begin with the left hand, and gradually transform it into the right hand, justifying the steps with the hypotheses that we mentioned.

\begin{align*}
\cap_{i=1}^{n+1} A_i &= (\cap_{i=1}^{n} A_i) \cap A_{n+1} 
& \text{by associativity of $\cap$} \\
&= A_n \cap A_{n+1} & \text{by induction hypothesis} \\
&= A_{n+1} & \text{since $A_{n+1} \subset A_n$, we use base case} 
\end{align*}

#### 1.2 $\cup_{i=1}^n A_i = A_1$

We will use same techniques as previous sub-exercise, but this time we will use more natural language instead of only mathematical notation. We will also omit some of the details, to make the proof briefer.

###### 1.2.1 Base case $n=2$

For $n=2$ the hyphothesis simplifies to:
\begin{equation*}
A_2 \subset A_1
\end{equation*}

And the equality to prove simplifies to 

\begin{equation*}
A_1 \cup A_2 = A_1
\end{equation*}

Let us argument first why $(A_1 \cup A_2) \subset A_1$: let $x \in A_1 \cup A_2$; by definition of the set operator $\cup$, the element $x$ belongs either to $A_1$, to $A_2$ or to both. If $x$ belongs to $A_1$ or to both, then the result holds immediately: the $x$ belongs to $A_1$. The remaining scenario is when $x \in A_2$, and in that case we use the hipothesis $A_2 \subset A_1$. This tells that any element of $A_2$ is also an element of $A_1$; hence $x \in A_1$, which is what we wanted to prove.

Let us argument now why $A_1 \subset (A_1 \cup A_2)$: let $y \in A_1$, by definition of set operator $\cup$, the element also belongs to $A_1 \cup A_2$ (this is actually a propositional logic result: $p \implies (p \lor q))$. And that's it, having $y \in A_1 \cup A_2$ is what we wanted to prove.

Together the two arguments above show that the sets are equal ($A_1 \cup A_2 = A_1$), and this concludes our base case.


###### 1.2.2 Inductive case

The property for $n+1$ in this case is:

\begin{equation*}
\cup_{i=1}^{n+1} A_i = A_1
\end{equation*}

Let us prove it again by starting with the left side, and gradually transform until we reach the right side (justifying each step):

\begin{align*}
\cup_{i=1}^{n+1} A_i &= A_1 \cup (\cup_{i=2}^{n+1} A_i) 
& \text{by associativity of $\cup$} \\
&= A_1 \cup A_2 & \text{by induction hypothesis (right side has $n$ elements)} \\
&= A_1 & \text{since $A_2 \subset A_1$, we use base case} 
\end{align*}


# 2
<b>If $X$ is the number of successes in $N$ Bernoulli trials, find the probability that $X >= 3$ given that $X >= 1$</b>

Let us assign names to our events: 

<ul>
<li>A := $X >= 3$</li>
<li>B := $X >= 1$</li>
</ul>

Then this question is asking for the conditional probability $P(A \mid B)$. We know that the definition of such is:

\begin{equation*}
P(A \mid B) = \dfrac{P(A \cap B)}{P(B)}
\end{equation*}

Now, we can leverage the previous exercise's result because $A \subset B$:

\begin{equation*}
A \subset B \;\implies\; A \cap B = A
\end{equation*}

Then our conditional probability simplifies to:

\begin{equation*}
P(A \mid B) = \dfrac{P(A)}{P(B)}
\end{equation*}

All we need to do now, is to give a more concrete value to $P(A)$ and $P(B)$. Let us begin with $P(A)$:

\begin{align*}
P(A) &= P(X >= 3) \\
     &= 1 - P(X < 3) \\
     &= 1 - \left[ P(X=0) + P(X=1) + P(X=2) \right] \\
     &= 1 - P(X=0) - P(X=1) - P(X=2)
\end{align*}

Similarly we proceed for $P(B)$:

\begin{align*}
P(B) &= P(X >= 1) \\
     &= 1 - P(X < 1) \\
     &= 1 - P(X=0)
\end{align*}

Thus, $P(A \mid B)$ becomes:

\begin{equation*}
P(A \mid B) = \dfrac{P(X >= 3)}{P(X >= 1)} = 
\dfrac{1 - P(X=2) - P(X=1) - P(X=0)}{1 - P(X=0)}
\end{equation*}

But given these are $n$ Bernoulli trials, we actually know the definition of $P(X=k)$:

\begin{equation*}
P(X=k) = C^n_k p^k q^{n-k}
\end{equation*}

Let us apply this definition to our terms $P(X=k)$:

\begin{align*}
P(X=0) &= C^n_0 p^0 q^{n-0} = q^n \\
P(X=1) &= C^n_1 p^1 q^{n-1} = n p q^{n-1} \\
P(X=2) &= C^n_2 p^2 q^{n-2} 
\end{align*}

Let us substitute now in our formula for $P(A|B)$ to yield the final result (alright, let us actually summarize the whole development):

\begin{align*}
P(A \mid B) 
&= \dfrac{P(A \cap B)}{P(B)} \\\\
&= \dfrac{P(\{X >= 3\} \cap \{X >= 1\})}{P(X >= 1)} \\\\
&= \dfrac{P(X >= 3)}{P(X >= 1)} \\\\ 
&= \dfrac{1 - P(X=2) - P(X=1) - P(X=0)}{1 - P(X=0)} \\\\
&= \dfrac{1 - C^n_2 p^2 q^{n-2} - n p q^{n-1} - q^n}{1 -q^n} 
\end{align*}

<b>Let $C_1$ be an unbiased coin, and $C_2$ a biased coin with probability of heads $\frac{3}{4}$. At time $t = 0$, $C_1$ is tossed. If the result is heads, then $C_1$ is tossed at time $t = 1$. If the result is tails, $C_2$ is tossed at $t = 1$. Then the process is repeated at time $t = 2, 3, \dots$. In general, if heads appears at $t = n$, then $C_1$ is tossed at time $t = n+1$. If tails appears at $t = n$, then $C_2$ is tossed at $t = n+1$.</b>

<b>Find $y_n$ = the probability that the toss at $t = n$ will be head (hint: use a difference equation)</b>

We can use the law of total probability, at time $n$. What occurs there, depends on what happened on time $n-1$ (aka $y_{n-1}$), but split in two disjoint cases. This is because depending on whether we got head or tail a time $n-1$, we used either coin $C_1$ or $C_2$. And for each coin, we know that the conditional probability of getting head is different. Thus, the recurrence relation can be explained as follows:

At time $n$, the probability of getting head is given by the sum of two probabilities: the probability of getting head at time $n-1$ times the probability of getting head on $C_1$; plus the probability of getting tail at time $n-1$ times the probability of getting head on $C_2$.

Let $H$ be the probability of getting head, our recurrence equation can be written formally as:

\begin{align*}
y_n &= y_{n-1} P(H \mid C_1) + (1 - y_{n-1})P(H \mid C_2) \\
    &= y_{n-1} \frac{1}{2} + (1 - y_{n-1})\frac{3}{4} \\
    &= y_{n-1} \frac{1}{2}  - y_{n-1}\frac{3}{4} + \frac{3}{4} \\
    &= - \frac{y_{n-1}}{4} + \frac{3}{4} \\
\end{align*}

Our recurrence relation can actually be seen as a recurrence equation:

\begin{equation*}
y_n + \frac{y_{n-1}}{4} = \frac{3}{4} 
\end{equation*}

Recurrence equations are the discrete counterpart of differencial equations, and this particular specimen is an inhomogeneus first order linear difference equation. The general recipe for solving those is as follows:

<ol>
<li>Find a particular solution for inhomogeneous equation.</li>
<li>Find the general solution for the associated homogeneous equation.</li>
<li>The general solution for inhomogeneous equation, is the sum of the two above.</li>
<li>Using initial conditions we can resolve parameters.</li>
</ol>

Let us begin by finding a particular solution for the inhomogeneous equation. Given that the right hand is constant, we can assume the particular solution is constant as well. This means that:

\begin{equation*}
y_{n} = y_{n-1} = c
\end{equation*}

Replacing the constant on the difference equation gives us:

\begin{align*}
c + \frac{c}{4} &= \frac{3}{4} \implies \\
\frac{5c}{4} &= \frac{3}{4} \implies \\
5c &= 3 \implies \\
c &= \frac{3}{5} 
\end{align*}

Then, our particular solution is constant $y_n = \frac{3}{5}$. Let us find now the general solution for the homogeneous equation:

\begin{equation*}
y_n + \frac{y_{n-1}}{4} = 0
\end{equation*}

The generic solution for such equation is given by:

\begin{equation*}
y_n = K \left(-\frac{1}{4}\right)^n
\end{equation*}

And thus, adding both together, leads to general solution for the inhomogeneus equation:

\begin{equation*}
y_n = K \left(-\frac{1}{4}\right)^n + \frac{3}{5}
\end{equation*}

Where the constant $K$ can be known by using initial conditions. We know what is $y_0$, because the problem says that at $t = 0$ we toss coin $C_1$; and thus:

\begin{equation*}
y_0 = P(H \mid C_1) = \frac{1}{2}
\end{equation*}

Replacing above value for $y_0$ into general solution, allows to find what is $K$:

\begin{align*}
y_0 &= K \left(-\frac{1}{4}\right)^0 + \frac{3}{5} \implies \\
\frac{1}{2} &= K + \frac{3}{5} \implies \\
K  &= \frac{1}{2} - \frac{3}{5} = -\frac{1}{10}
\end{align*}

Therefore, the final solution for inhomogeneous equation would be the following:

\begin{equation*}
y_n = -\frac{1}{10} \left(-\frac{1}{4}\right)^n + \frac{3}{5}
\end{equation*}

which is the probability we were asked for.

<b>An employer is about to hire one new employee from a group of $N$ candidates, whose future potential can be rated on a scale from $1$ to $N$. The employer proceeds according to the following rules:</b>

<b>(a) Each candidate is seen in succession (in random order) and a decision is made whether to hire the candidate.</b>

<b>(b) Having rejected $m-1$ candidates ($m > 1$), the employer can hire the $m_{th}$ candidate only if the $m_{th}$ candidate is better than the previous $m-1$.</b>

<b>Suppose a candidate is hired on the $i_{th}$ trial. What is the probability that the best candidate was hired?</b>

The rule (a) pretty much tells us that the candidates are to be taken from a shuffled array. The resulting random permutation could be any, and we know that a hiring decision was made at position $i$ (our index starts at $1$). 

The rule (b) tells us that the previous $i-1$ candidates had a lower ranking, so we can conclude that best guy was not among them. This means that the best candidate must have an index from the set:

\begin{equation*}
\{ i, i+1, i+2, \cdots, N \}
\end{equation*}

The size of the previous set is $N - (i-1)$, because we are discounting the rejected $i-1$ folks who are worse than candidate $i$. Thus, from this remaining set any guy could be the best; that is, the probability that any of those indexes to correspond to best candidate is:

\begin{equation*}
P(\text{any remaining index to be the best}) = 
\frac{1}{N-(i-1)} = 
\frac{1}{N-i+1}
\end{equation*}

Above reasoning applies in particular to the chosen candidate $i$, hence the probability of having hired the best candidate is given by:

\begin{equation*}
P(\text{best candidate being at index } i) = 
\frac{1}{N-i+1}
\end{equation*}
